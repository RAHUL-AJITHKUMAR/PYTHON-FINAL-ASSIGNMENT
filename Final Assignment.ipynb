# Distrubtuion of employees across team and percentage split 

```python
# Number of employees per team
team_distribution = df["Team"].value_counts()

# percentage split
team_percentage = (team_distribution / len(df)) * 100

# DataFrame
team_summary = pd.DataFrame({"Employee Count": team_distribution, "Percentage": team_percentage})


plt.figure(figsize=(10,10))
plt.pie(team_summary["Employee Count"], labels=team_summary.index, autopct="%1.1f%%", colors=plt.cm.Paired.colors)
plt.title("Percentage Split of Employees Across Teams")
plt.show()

team_summary
```

# Which age group is Predominant ??

```python
# bins and labels
bins = [18, 25, 30, 35, 40, 50, 60]  # Age group ranges
labels = ["18-25", "26-30", "31-35", "36-40", "41-50", "51-60"]  # Labels for age groups

# Categorize employees into age groups
df["Age Group"] = pd.cut(df["Age"], bins=bins, labels=labels, right=True)

# Count the number of employees in each age group
age_group_distribution = df["Age Group"].value_counts().sort_index()

# Convert to DataFrame for better presentation
age_group_df = pd.DataFrame({ "Age Group": age_group_distribution.index,"Employee Count": age_group_distribution.values})


# Predominant age group
predominant_age_group = age_group_distribution.idxmax()
print(f"\nPredominant Age Group: {predominant_age_group}")

age_group_df
```

# Team and Position with Highest salary

```python
# highest salary - team & position
highest_team = df.groupby("Team")["Salary"].sum().idxmax()
highest_team_amount = df.groupby("Team")["Salary"].sum().max()
highest_position = df.groupby("Position")["Salary"].sum().idxmax()
highest_position_amount = df.groupby("Position")["Salary"].sum().max()

# Print results
print(f"Team with Highest Salary Expenditure: {highest_team} - ${highest_team_amount:,.2f}")
print(f"Position with Highest Salary Expenditure: {highest_position} - ${highest_position_amount:,.2f}")
```

# Finding out Correlation between Age and Salary

```python
correlation = df["Age"].corr(df["Salary"])
print(f"Correlation between Age and Salary: {correlation:.2f}")

import plotly.express as px
fig = px.scatter(df, x="Age", y="Salary", 
                 title="Interactive Correlation Between Age and Salary",
                 labels={"Age": "Employee Age", "Salary": "Employee Salary ($)"},
                 trendline="ols",  # Add trendline for correlation
                 opacity=0.6, color=df["Age"], color_continuous_scale="Viridis")

fig.show()
```
